```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
from sklearn.datasets import load_boston
```

```python
boston = load_boston() # 讀入
```

```python
boston.feature_names # 展示featrues
```

```python
X = boston.data
Y = boston.target
plt.plot(X, Y)
```

```python
len(X)
```

```python
len(Y)
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.5, random_state = 13)
```

```python
x_train
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
regr = LinearRegression()
```

```python
x_train.shape
```

```python
regr.fit(x_train, y_train)
```

```python
y_predict = regr.predict(x_test)
```

```python
plt.scatter(y_predict, y_test)
plt.plot([0, 50], [0, 50], 'r') # 畫出基準(==對角線)
plt.xlabel("Predicted price")
plt.ylabel("True price")
```

### 技巧2: 畫多張圖

```python
x = np.linspace(1, 50, 50)
```

```python
plt.figure(figsize=(8, 10))
plt.subplot(2, 2, 1) # (幾row, 幾column, 第幾張圖)
plt.plot(x)
plt.title('x')

plt.subplot(2, 2, 2)
plt.plot(x**2)
plt.title('x**2')

plt.subplot(2, 2, 3)
plt.plot(x**(1/2))
plt.title('x**(1/2)')

plt.subplot(2, 2, 4)
plt.plot(x**(-2))
plt.title('x**(-2)')
```

```python
plt.figure(figsize=(16, 26))
for i, feature in enumerate(boston.feature_names):
    plt.subplot(5, 3, i+1)
    plt.scatter(X[ : , i], Y, s=10) # X[ : , i]: 在第一個維(index)中，取全部的數據；在第二個維(feature)中，取第i個數據；Y: X[index, feature]所對應到的值
    plt.ylabel("price", fontsize=15)
    plt.xlabel(feature, fontsize=15)
#     plt.tight_layout()
```
