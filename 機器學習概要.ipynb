```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

# 分類

```python
x = np.array([[1, -5], [6, 1], [-7, 4], [-3, -6]])
y = np.array([1, 2, 2, 1])
```

```python
plt.scatter(x[:, 0], x[:, 1], s=50, c=y) # scatter(x座標, y座標, s=點的大小, c=如何上色/分類)
```

```python
from sklearn.svm import SVC
# SVM: Support Vector Mechine 支持向量機
# SVC: Support Vector Classifier 支持向量分類器
```

```python
clf = SVC() # 打開機器
```

```python
clf.fit(x, y) # fit(輸入資料, 答案)
```

```python
clf.predict([[6, 1]])
```

```python
clf.predict([[0, 0]])
```

```python
clf.predict(x)
```

```python
plt.scatter(x[:, 0], x[:, 1], s=50, c=y)
plt.plot(clf.predict(x))
# 錯誤示範
```

## 分類全版格子點

### meshgrid (+linspace): 畫格子點

```python
ex3 = [1, 2, 3, 4]
ex4 = [5, 6]
ex5 = [9]
ex6 = [10, 11, 12]
np.meshgrid(ex3, ex4, ex5, ex6) # meshgrid(第一維(最外圈)重複幾次, 第二維重複幾次, 第三維重複幾次, 第n維重複幾次...)
```

```python
EX3, EX4 = np.meshgrid(ex3, ex4)
EX3, EX4
```

```python
plt.scatter(EX3, EX4)
```

```python
EX3, EX4, EX5 = np.meshgrid(ex3, ex4, ex5)
EX3, EX4, EX5
```

```python
plt.scatter(EX3, EX4, EX5)
```

```python
X, Y = np.meshgrid(np.linspace(-8, 8, 30), np.linspace(-11, 0, 30))
```

```python
plt.scatter(X, Y)
```

```python
X
```

```python
Y
```

### 把二維陣列拉平成一維: ravel

```python
X = X.ravel()
Y = Y.ravel()
```

```python
X
```

```python
Y
```

### zip: 將兩個一維陣列的各項分別合併

```python
ex1 = [1, 2, 3, 4]
ex2 = [5, 6, 7, 8]
list(zip(ex1, ex2))
```

```python
list(zip(X, Y))
```

```python
Z = clf.predict(list(zip(X, Y))) # 以何為界? 請參考clf.fit(x, y)那裡
Z
```

```python
plt.scatter(X, Y, c=Z) # 如此一來，X, Y, Z(classifier)皆同維(一維)
```

## 用KMeans分類

```python
X = np.random.rand(100, 3) # rand(n筆資料, 每筆有m個數字)
X = X * 100
X
```

```python
plt.scatter(X[:, 0], X[:, 1], s=50)
```

```python
from sklearn.cluster import KMeans
```

```python
# 開KMeans機器
clf = KMeans(n_clusters=3) # KMeans(n_clusters=你要分幾群)
```

```python
clf.fit(X)
```

```python
clf.labels_ # 看clf的labels
```

```python
 plt.scatter(X[:, 0], X[:, 1], c=clf.labels_)
```
