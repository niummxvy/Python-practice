## 自製Dataframe

```python
score1 = np.random.randn(12, 3)
score1 = np.around(score1, decimals=2) # 四捨五入
score1
```

```python
epi1 = []
for i in range(1, 13):
    fmt = "第{}集".format(i)
    epi1.append(fmt)
# epi = ["第1集", "第2集", "第3集", "第4集", "第5集", "第6集", "第7集", "第8集", "第9集", "第10集", "第11集", "第12集"]
df1 = pd.DataFrame(score1, columns=["佩可", "凱留", "可可蘿"], index=epi1) # index: 可自訂index範圍
df1
```

```python
score2 = np.random.randn(18, 3)
score2 = np.around(score2, decimals=2)
score2
```

```python
EP = []
for i in range(10):
    fmt = "第{}集".format(i)
    EP.append(fmt)
EP
```

```python
epi2 = []
for i in range(1, 19):
    fmt = "第{}集".format(i)
    epi2.append(fmt)
df2 = pd.DataFrame(score2, columns=["佩可", "凱留", "可可蘿"], index=epi2)
df2
```

### 合併(concat)

```python
df3 = pd.concat([df1, df2], axis=0) # axis=0: df2接在df1最後一row下方；axis=1: df2接在df1最後一col右方；
df3
```

```python
epi = []
for i in range(1, 31):
    fmt = "第{}集".format(i)
    epi.append(fmt)
df3.index = epi
df3
```

```python
df4 = pd.concat([df1, df2], axis=1) # axis=0: df2接在df1最後一row下方；axis=1: df2接在df1最後一col右方；
df4
#前3 col為df1，後3 col為df2
```
